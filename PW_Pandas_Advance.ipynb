```python
# find how many pasangers who have paid more than average
print(df['Fare'].mean())
```

# how to apply comdition on more than 2 columns
df[(condition1) & (condition2)]

# For more such questions see
https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html

```python
# find how many males were traveling in class 1
print(len(df[(df['Sex'] == 'male') & (df['Pclass'] == 1)]))                               # here we have applied condition on 2 columns
```

```python
# find how many people survived in the ship
df[df['Survived'] == 1]
```

```python
# find name of females who have paid more than average
df[(df['Sex'] == 'female') & (df['Fare'] > df['Fare'].mean())]['Name']
```

```python
# find the name of the male who have paid max fair and how much
df[(df['Sex'] == 'male') & (df['Fare'] == df['Fare'].max())] [['Name', 'Fare']]
```

# How to take out the data at a paticular intersection of row and column

whenever using iloc or loc 1st row will come then column

https://towardsdatascience.com/how-to-use-loc-and-iloc-for-selecting-data-in-pandas-bd09cb4c3d79

```python
df.iloc[0:5 , 1:7]               # in iloc will take inbuilt index by default index or index number
```

```python
# can also bes use by writing the column name as well
df.loc[0:5 , ['Pclass', 'Name', 'Sex']]                            # loc will take name index
```
