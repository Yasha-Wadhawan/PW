# functions in numpy which helps to generate array, helps in generating different dimentions and types of array

```python
# in fromfunction() pass any function (def or lambda) and pass the dimention of the array to make the array
# in this the function passed as the parameter takes the rank(position) of the rows and column  as a default parameters
# https://www.educative.io/answers/what-is-fromfunction-from-numpy-in-python

def array(i,j):
  return i==j

np.fromfunction(array , (3,3))

# np.fromfunction(lambda i,j : i==j , (3,3))

```

```python
def array2(i,j):
  return i>j

np.fromfunction(array2 , (4,4))
```

```python
def array3(i,j):
  return i+j

np.fromfunction(array3 , (3,3))                              # here the default parameters of the array3 are the rank(index) in the rows and column ,,, 1st row then column
```

```python
np.fromfunction(lambda i,j : i*j , (5,3))
```

```python
# in fromiter function  any iterable object can be passed and can be converted into 1d array
# iterable objects like any list, tuple,  dictionary can be passed  (will not work with string)
# in pw they have passed a list comprehension

# https://www.educative.io/answers/what-is-the-numpyfromiter-function-in-python
# https://chat.openai.com/share/889f6126-37c6-48eb-b9a5-75bf16e7daca

#https://www.programiz.com/python-programming/list-comprehension

iterable = (i*i for i in range(5))                               # this way of making list is called list comprehension

np.fromiter(iterable, float)
```

# can also generate random samples

```python
# generate random normal distribution

# Setting up our mean and standard deviation
mean, std_deviation = 0, 0.1 # mean and standard deviation
# Applying the numpy random normal method
random_normal = np.random.normal(mean, std_deviation, 100)
print(random_normal)

```

```python
# Generate a Random Uniform Distribution Data

# Using numpy random uniform to create random samples
sample = np.random.uniform(-1,0,1000)
# Uniform Distribution check
print(np.all(sample >= -1))
print(np.all(sample < 0))
```

# indexing and slicing in array

```python
import pandas as pd
import numpy as np

# Define the number of rows
num_rows = 777

# Create a dictionary to store the data
data = {
    "Student_Name": ["Student {}".format(i) for i in range(1, num_rows + 1)],
    "Math_Score": np.random.randint(50, 100, num_rows),
    "English_Score": np.random.randint(50, 100, num_rows),
    "History_Score": np.random.randint(50, 100, num_rows),
    "Attendance": np.random.randint(70, 100, num_rows),
    "Study_Hours": np.random.randint(2, 8, num_rows),
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("student_scores.csv", index=False)
```
