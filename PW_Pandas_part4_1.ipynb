# Join operation

join is very similar to merge the only difference is merge is column based and join in index based

in merge how is inner and in join how is left by default

```python
data3 = pd.DataFrame({ 'key1':[1,2,3,4,5],
     'key2' : [6,7,8,9,10],
      'key3' : [11,12,13,14,15],
      'key4' :[16,17,18,19,20]},
       index = ['a','b','c','d','f']
                     )
```

```python
data4 = pd.DataFrame({ 'key8':[1,27,3,46,5],                         # call columns name are kept diferent but 2 index anme is same from data3
     'key5' : [6,74,8,98,10],
      'key6' : [114,12,163,14,15],
      'key7' :[16,197,18,819,20]},
       index = ['a','b','h','i','j']
                     )
```

```python
data3
```

```python
data4
```

```python
data3.join(data4)                                    # in join by default how = 'Left'
```

```python
data3.join(data4, how='right')
```

```python
data3.join(data4, how='inner')
```

```python
data3.join(data4, how='outer')
```

```python
data3.join(data4, how='cross')
```

# make a catagorical column which consist of if fare being divided into 3 categories that is cheap, medium, high

```python
def fare_cato(x):
  if x<=10:
    return 'cheap'
  elif x<=50:
    return 'medium'
  else:
    return 'high'

df['cate_fare'] = df['Fare'].apply(fare_cato)
```
